<h1>Test Cricket Data Analysis</h1>
<br>goal of this notebook is to analyze test cricket data in given dataset

### Import required libraries

```python
import numpy as np
import pandas as pd
```

### Reading xlsx file

```python
df = pd.read_excel("test_cricket.xlsx", engine='openpyxl' , sheet_name='wickets')

display(df.head(10))
```

### Description of the features :

<br><strong>Player: </strong> the name and country name of the player
<br><strong>Span: </strong> duration of carrer
<br><strong>Mat: </strong> number of matches played
<br><strong>Inns: </strong> number of innings played
<br><strong>Balls: </strong> number of balls bowled 
<br><strong>Runs: </strong> number of runs
<br><strong>Wkts: </strong> number of wickets taken
<br><strong>BBI: </strong> best bowling in innings
<br><strong>BBM: </strong> best bowling in match
<br><strong>Ave: </strong> average runs per wicket
<br><strong>Econ: </strong> average runs concided per over
<br><strong>SR: </strong> strike rate
<br><strong>5: </strong> number of times 5 wickets taken in match
<br><strong>10: </strong> number of times 10 wickets taken in match

### Renaming columns in dataset

```python
#printing columns in dataframe
print(df.columns)
```

```python
#renameing columns in dataframe
df = df.rename(columns={'Span' : 'Carrer_Duration',
                        'Mat' : 'Match',
                        'Inns': 'Innings',
                        'Wkts' : 'Wickets',
                        'BBI' : 'Best_Bowling_Innings',
                        'BBM' : 'Best_Bowling_Match',
                        'Ave' : 'Bowling_Average',
                        'Econ': 'Economy_Rate',
                        'SR' : 'Strike_Rate'})
display(df.head())
```

In the dataframe we can see that the name of the player and the country of the player is in same column lets seperate that

### Spliting and Repositioning the columns

```python
#splitting player into two columns
df_player = df['Player'].str.split("(" , expand = True)

display(df_player.head())
```

```python
#concating df_player to main dataframe
df = pd.concat([df,df_player], axis=1)

display(df.head())
```

```python
#removing old player column
df = df.drop('Player' , axis=1)

display(df.head())
```

```python
#renameing new player and country column

df = df.rename(columns={0: 'Player',
                        1: 'Country'})

display(df.head())
```

```python
#removing ')' in country column

df['Country'] = df['Country'].str.replace(")", "")

display(df.head())
```

```python
#rearranging column sequence

new_col_seq = ['Player' , 'Country' , 'Carrer_Duration' , 'Match' , 'Innings' , 'Balls' , 'Runs' , 'Wickets' , 'Best_Bowling_Innings' , 'Best_Bowling_Match' , 'Bowling_Average' , 'Economy_Rate', 'Strike_Rate', 5 ,
 10]

df = df[new_col_seq] 

display(df.head())
```
